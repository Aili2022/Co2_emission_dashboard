```python
import pandas as pd
import numpy as np
import panel as pn

pn.extension('tabulator') # to create interactive tables

import hvplot.pandas # to create interactive dataframes
```

```python
data_file = '/Users/aili/Documents/Self-Learning/co2_dashboard/owid-co2-data.csv'
df = pd.read_csv(data_file)
```

```python
df[df['country'] == 'World']
```

```python
# minior data preprocessing
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/ df['population'], 0)
```

```python
# Make Dataframe Pipeline Interactive
idf = df.interactive() #calling the method on the dataframe object
```

```python
#Co2 emission over time by continent
# Define Panel Widgets
year_slider = pn.widgets.IntSlider(name = 'Year slider', start = 1750, end = 2023, step = 5, value = 1850) 
#interactive elements that user to configuare the visualization
year_slider
```

```python
# Radio buttons for Co2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = "Y axis",
    options = ['co2','co2 per capita'],
    button_type = 'success'
)
```

```python
continents = ['Wourld', 'Asia','Oceania','Europe', 'Africa', 'NA', 'SA', 'Antarctica']

co2_pipeline = (
    idf[
    (idf.year <= year_slider) &
    (idf.country.isin(continents))
    ]
    .groupby(['country','year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by = 'year')
    .reset_index(drop = True)
)
```

```python
co2_plot = co2_pipeline.hvplot(x = 'year', by = 'country', y = yaxis_co2, line_width = 2 ,title = "Co2 emiision by continent")
co2_plot
```

```python
#Tabel -Co2 emission over time by continent
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination = 'remote', page_size = 10, sizing_mode = 'stretch_width')
co2_table
```

```python
#Co2 vs GDP scatterplot
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~(idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by = 'year')
    .reset_index(drop = True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline 
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x = 'gdp_per_capita', 
                                                                         by = 'country',
                                                                         y = 'co2', 
                                                                         size=80, 
                                                                         kind="scatter",
                                                                         alpha=0.7,
                                                                         legend=False,
                                                                         height=500,
                                                                         width=500
                                                                        )
```

```python
co2_vs_gdp_scatterplot
```

```python
#Last step to create all the graph together into a dashborad
#Layout using Template
template = pn.template.FastListTemplate(
    title = 'An clear review of World Co2 emission dashboard',
    sidebar = [pn.pane.Markdown("# Co2 Emssions and Climate Change"),
               pn.pane.Markdown("## Settings"),
               year_slider],

    main = [pn.Row(pn.Column(yaxis_co2,
                         co2_plot.panel(width=700),
                         margin=(0, 25)),
               co2_table.panel(width=500)),
            pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=600),
                         margin=(0, 25)))],
    
    accent_base_color = "#88d8b0",
    header_background = "#88d8b0",
)

# show the template
template.servable();
```
